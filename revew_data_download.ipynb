```python
import os
import xmltodict
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
def extract_from_file(data):
    
    for _data in (f"{d}</review>" for d in data.split("</review>")):
        bs_data = BeautifulSoup(_data, "xml")
        if bs_data.review_text != None:
            yield bs_data.review_text.text.replace("\n","")
        
def assign_class(dataset, data, category, index_start=0):
    
    for i, text in enumerate(extract_from_file(data), index_start):
        dataset[i] = {"category": category, "text": text}

    return dataset, i
```

```python
rel_path = 'nlp_data_reviews'

dataset, i = {}, 0

for category in ["book","dvd","electronics","kitchen_houseware"]:
    with open(os.path.join(rel_path,f"{category}_review.xml")) as f:
        dataset, i = assign_class(dataset, f.read(), category, i)
```

```python
with open("review_data.json", "w") as file:
    json.dump(dataset, file)
```
